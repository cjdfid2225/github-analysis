**objective**

현재 Github은 우리가 web page로 접근가능한 대부분의 정보들을 API로도 받을 수 있게 설계되어 있음. Gitub이 제공하는 API를 바탕으로 필요한 데이터들을 어떻게 취득할 수 있는지를 확인하는 것을 목표로 함


```python
import os
import json
import requests
import pandas as pd

# credential이 필요
GITHUB_URL = "https://api.github.com"
GITHUB_GQL = "https://api.github.com/graphql"

# reference : https://github.com/settings/tokens
GITHUB_KEY = "../credentials/github.key"
if os.path.exists(GITHUB_KEY):
    with open(GITHUB_KEY, 'r') as f:
        api_key = f.readline()        
        headers = {"Authorization": "bearer " + api_key}
```

# Github API 탐색하기


## 1. event 조회하기

깃헙 최근 public 활동 30개를 가져옴

* [reference](https://docs.github.com/en/rest/reference/activity)

#### RestFUL API 방식

```python
res = requests.get(
    "/".join((GITHUB_URL, 'events')),
    headers=headers)
if res.status_code == 200:
    print(json.dumps(json.loads(res.text)[:1],indent=2))
```

## 2.  Organization 조회하기

### (1) 전체 Organization 가져오기

```python
res = requests.get(
    "/".join((GITHUB_URL, 'organizations')),
    headers=headers)
if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```

### (2) 특정 Organization 가져오기

#### RestFUL API 방식

```python
res = requests.get(
    "/".join((GITHUB_URL, 'orgs','tensorflow')),
    headers=headers)
if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```

#### GraphQL API 방식

```python
query = """
query GetOrgs($login: String!) { 
  organization(login: $login) { 
    description,
    name,
    avatarUrl,
    repositories(first: 100){
      nodes {
        name
      }
    }    
  }
}
"""
variables = {
    "login": "tensorflow"}
res = requests.post(GITHUB_GQL, headers=headers,
                    json={'query': query , 'variables': variables})

if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```

## 3. User 조회하기

* [Reference](https://docs.github.com/en/rest/reference/users#list-users)

### (1) 전체 user 가져오기 

```python
res = requests.get(
    "/".join((GITHUB_URL, 'users')), 
    params={"since":30})
if res.status_code == 200:
    df = pd.DataFrame(res.json())

df.head()
```

### (2) 특정 user 가져오기

#### RestFUL API 방식

```python
res = requests.get(
    "/".join((GITHUB_URL, 'users','craftsangjae')),
    headers=headers)
if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```

#### GraphQL API 방식

```python
query = """
query GetUser($login: String!) {
  user(login: $login){
    bio
    company
    avatarUrl
    email
    followers(first:100) {
      edges {
        node {
          login
        }
      }
    }
    following(first:100) {
      edges {
        node {
          login
        }
      }
    }
  }
}
"""
variables = {
    "login": "craftsangjae"}
res = requests.post(GITHUB_GQL, headers=headers,
                    json={'query': query , 'variables': variables})

if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```

## 4. Repository 조회하기

### (1) 전체 Repository 조회하기

```python
res = requests.get(
    "/".join((GITHUB_URL, 'repositories')), 
    params={"since":30})
if res.status_code == 200:
    df = pd.DataFrame(res.json())

df.head()
```

### (2) 특정 Repository 가져오기

#### RestFUL API 방식

```python
res = requests.get("/".join((
    GITHUB_URL, 'repos', 'tensorflow','tensorflow')))

if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```

#### GraphQL API 방식

```python
query = """
query GetRepo($owner: String!, $name: String!) { 
  repository(owner:$owner, name:$name) { 
    forkCount,
    isFork,
    name,
	  owner {
      login
    },
    stargazers(first:10){
      nodes{
        name
      }
    },
    watchers(first:10){
      nodes{
        name
      }
    },
    languages(first:10) {
      nodes {
        name
      }
    },
    createdAt,
    updatedAt,
    hasWikiEnabled,
    hasIssuesEnabled,
    hasProjectsEnabled,
    viewerHasStarred,
    description,
    repositoryTopics(first:10){
      nodes {
        topic{
          name
        }
      }
    }
  }
}
"""
variables = {"owner": "tensorflow",
             "name": "tensorflow"}


res = requests.post(GITHUB_GQL, headers=headers,
                    json={'query': query , 'variables': variables})

if res.status_code == 200:
    print(json.dumps(json.loads(res.text),indent=2))
```
